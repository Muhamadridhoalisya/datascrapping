```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import re

# Base URL website
base_url = "https://peraturan.go.id/uu?page="

# Loop through all pages (1 to 88)
for page_num in range(1, 89):
    # Construct the URL for the current page
    url = f"{base_url}{page_num}"
    
    # Membuat permintaan GET ke URL
    response = requests.get(url)
    html_content = response.content
    
    # Membuat objek BeautifulSoup dari respons HTML
    soup = BeautifulSoup(html_content, "html.parser")
    
    # Mencari semua elemen yang mengandung tautan ke file PDF
    strip_elements = soup.find_all("div", class_="strip grid")
    
    # Iterasi melalui setiap elemen dan unduh file PDF
    for strip in strip_elements:
        # Mendapatkan nama file dari elemen <p>
        p_element = strip.find("p", style=re.compile(r"padding-top: -2"))
        if p_element:
            pdf_filename = p_element.get_text(strip=True) + ".pdf"
            
            # Mencari tautan PDF dalam elemen tersebut
            pdf_link = strip.find("a", href=lambda href: href and href.endswith(".pdf"))
            if pdf_link:
                pdf_url = pdf_link.get("href")
                pdf_url = urljoin(url, pdf_url)
                
                # Membuat permintaan GET ke URL PDF dengan header
                pdf_response = requests.get(pdf_url, headers={'User-Agent': 'Mozilla/5.0'})
                
                # Verifikasi respons HTTP
                if pdf_response.status_code == 200:
                    # Menyimpan file PDF
                    with open(pdf_filename, "wb") as pdf_file:
                        pdf_file.write(pdf_response.content)
                    
                    print(f"File PDF {pdf_filename} berhasil diunduh.")
                else:
                    print(f"Gagal mengunduh {pdf_url}, status code: {pdf_response.status_code}")
```
